```python
'''  For making computation faster, first let joins appropriate months into quarters  '''
quarterly_data=[]
for i in range(int(len(list_of_DFs)/3)):
    quarterly_data.append(pd.concat(list_of_DFs[i*3:i*3+3]))
    if i%500==0:
        print(i)
        
for i in range(int(len(list_of_DFs2)/3)):
    quarterly_data.append(pd.concat(list_of_DFs2[i*3:i*3+3]))
    if i%500==0:
        print(i)
```

```python
''' Function for transformation of strings to floats; missing values replaced by np.NaN ''' 
def trans(table,i):
    for j in range(len(table.iloc[:,i])):
        y=table.iloc[j,i].split(".")
        z=".".join([re.sub("[^0-9]"," ",x)  for x in y])
        try:
            table.iloc[j,i]=float(z)
        except:
            table.iloc[j,i]=np.NaN
```

```python
'''  Converting strings to numbers where needed, for time efficiency step by step  '''
for t in range(len(first_q)):
    if t%100==0:
        print(t)
    table=first_q[t]
    for i in range(4,10):
        trans(table,i)
```
