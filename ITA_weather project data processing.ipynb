```python
'''  For making computation faster, first let join appropriate months into quarters  '''
quarterly_data=[]
for i in range(int(len(list_of_DFs)/3)):
    quarterly_data.append(pd.concat(list_of_DFs[i*3:i*3+3]))
    if i%500==0:
        print(i)
        
for i in range(int(len(list_of_DFs2)/3)):
    quarterly_data.append(pd.concat(list_of_DFs2[i*3:i*3+3]))
    if i%500==0:
        print(i)
```

```python
''' Making lists of quarterly dataframes, for convenience using copy of the original dataframe 
in order to have more recent reference ''' 
first_q=[x.copy(deep=True) for x in quarterly_data[::4]]
second_q=[x.copy(deep=True) for x in quarterly_data[1::4]]
third_q=[x.copy(deep=True) for x in quarterly_data[2::4]]
fourth_q=[x.copy(deep=True) for x in quarterly_data[3::4]]
```

## Transformation of the strings into floats and NaNs; for this purpose we use function "trans" together with function "applied_trans", which uses "apply" method to apply "trans" function to 5th-10th columns of each table in a quarter.

```python
import re
```

```python
def trans(item):
    y=item.split(".")
    y2=[x.split("-") for x in y]
    y=["-".join([re.sub("[^0-9]"," ",x)  for x in a]) for a in y2]
    z=".".join([x  for x in y])
    try:
        item=float(z)
    except:
        item=np.NaN
    return item
```

```python
def applied_trans(quarter):
    for t in range(len(quarter)):
        if t%500==0:
            print(t)
        for i in range(4,10):
            quarter[t].iloc[:,i]=quarter[t].iloc[:,i].apply(trans)
```

```python
'''  Converting strings to numbers where needed  '''
applied_trans(first_q)
```

```python
'''  Converting strings to numbers where needed  '''
applied_trans(second_q)
applied_trans(third_q)
applied_trans(fourth_q)
```
